## 로지스틱 회귀(Logistic Regression)
### 1. 개념
선형 모델을 분류에 사용해 샘플이 특정 클래스에 속할 확률을 추정하는데 사용할 수 있음. 반응변수가 범주형인 경우 적용하는 회귀분석을 로지스틱 회귀분석이라 함. 반응변수 Y를 직접 모델링하지 않고, Y가 특정 범주에 속하는 확률을 모델링함. 예를 들어 누군가의 카드 대금이 주어졌을 떄 그 사람이 카드대금을 연체할 확률을 구하는 것.
