```python
time_steps= [["1990","2000"],
             ["2000","2010"],
             ["2010","2015"],
             ["2015","2020"],
             ["2020","2022"],]

# 매직넘버 어떻게 정하나? 내가 안하고 prophet이 but 적당히 길고 적당히 짧아야 함. 너무짧거나 긴건 노가다 뛰어야함.
```

```python
fig, axes = plt.subplots(2,3)
fig.set_size_inches(16,9)
for i in range(5):
    ax = axes[i//3,i%3]
    df = samsung.loc[(samsung.index > time_steps[i][0]) & (samsung.index < time_steps[i][1])] #lable index
    # index => date
    sns.lineplot(x=df.index, y=df["Close"], ax=ax)
    ax.set_title(f"{time_steps[i][0]}~{time_steps[i][1]}")
    ax.set_xlabel("시간")
    ax.set_ylabel("가격")
plt.tight_layout()
plt.show()
```

```python
from prophet import Prophet
from prophet.plot import plot_plotly, plot_components_plotly

# fbprophet -> 안돌아감. 옛날버전.
```

* prophet에 들어가는 컬럼은 y컬럼과 ds컬럼이 존재해야 됨. ( y 예측값, ds 시간 )

```python
# 데이터 가공, 시계열은 na값 큰일남. 채워 넣어야함. fillna 일단 y ds 만든다.
stock = fdr.DataReader("005930","2022")
stock["y"]=stock["Close"]
stock["ds"]=stock.index
stock.head()
```

```python
# raw 데이터 피팅
m=Prophet()
m.fit(stock)
```

```python
future=m.make_future_dataframe(periods=30) #(periods: Any, freq: str = 'D', include_history: bool = True) -> DataFrame 날짜기준
future.tail()
#원하는 기간(D)을 정함 #ex 워킹데이 7일이면 9일 예측해야함.
```

```python
# 예측
forecast = m.predict(future) # yhat == 예측값
forecast.tail()
```

```python
# 
forecast[["ds","yhat","yhat_lower","yhat_upper"]].iloc[-40:-20] #loc아니고 iloc(순서기반) 인덱스주의! 시계열 데이터는 인덱스가 시간이므로 주의!!
# 잘 모르겠으면 loc 먼저 넣어보기 생각보다 iloc는 잘나옴.
```

```python
# 전체데이터 확인
m.plot(forecast) 
```

```python
# 이왕이면 plotly 사용하세요! 
fig = plot_plotly(m,forecast) # 전체를 다 보는것 # 올리면 정보가 뜬다!!!
fig
```

```python
# 세부 컴포넌트 이용해서 내용 확인
plot_components_plotly(m ,forecast)
```

```python
# 변경점 확인
from prophet.plot import add_changepoints_to_plot
fig=m.plot(forecast)
a= add_changepoints_to_plot(fig.gca(),m,forecast)
```

```python
# 유연성 => 계수가 낮으면 과소적합, 높으면 과대 적합 ==> 하이퍼 파라미터!! 사람이 한다!!!
m = Prophet(changepoint_prior_scale=0.8)
forecast = m.fit(stock).predict(future)
fig = m.plot(forecast)
```
